#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
a = np.arange(0,100)
b = np.exp(a)
c = a**2
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2) 
ax[0].plot(a,b, label = 'y = e**x')
ax[1].plot(a,c, label = 'y = x**2')
ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[1].set_xscale('linear') #por defecto lo que sale
ax[1].set_yscale('linear') #por defecto lo que sale
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
fig, ax = plt.subplots(nrows = 1, ncols = 2) 
ax[0].plot(a,b, label = 'y = e**x')
ax[1].plot(a,c, label = 'y = x**2')
ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[1].set_xscale('linear') #por defecto lo que sale
ax[1].set_yscale('linear') #por defecto lo que sale
ax[0].legend(loc="upper left")
ax[1].legend(loc="upper left")
plt.legend()
plt.show()
```
